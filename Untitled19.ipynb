```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('train.csv')

print("証拠品（データ）の読み込みが完了しました！")
print(train_df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='HeatingQC', y='SalePrice', data=train_df, order=['Ex', 'Gd', 'TA', 'Fa', 'Po'])
plt.title('Case 01: Heating Quality vs Sale Price')
plt.show()


print(train_df.groupby('HeatingQC')['SalePrice'].mean().sort_values(ascending=False))
```
