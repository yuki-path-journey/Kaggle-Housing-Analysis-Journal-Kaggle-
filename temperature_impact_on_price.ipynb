```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train_df = pd.read_csv('train.csv')
sns.histplot(train_df['SalePrice'], kde=True)
plt.title('Distribution of Sale Price')
plt.show()
```

Investigation Step 1: Distribution Analysis 住宅価格の分布を確認したところ、15万ドル付近に最も多く分布しているが、一部に50万ドルを超える「高額物件」が存在する。 次の任務: 暖房の品質（HeatingQC）や水捌けなどの要素が、この「高額物件」を作り出す真犯人なのかを特定する。

```python
print("--- 水捌け(LandSlope)の内訳 ---")
print(train_df['LandSlope'].value_counts())
sns.barplot(x='LandSlope', y='SalePrice', data=train_df)
plt.title('Land Slope vs Sale Price')
plt.show()
```

Case 02: Land Slope (水捌けと地形) [調査結果] 大半の物件は平坦な土地（Gtl）に位置するが、データ上は急傾斜地（Sev）の方が平均価格が高いという意外な結果が出た。 [考察] 傾斜地は水捌けの懸念があるものの、アメリカの不動産市場では「眺望（View）」が価格に強く反映されるケースがある。しかし、Sevはわずか13件とサンプルが少なく、特定の豪邸が平均を押し上げている「外れ値」の可能性も高い。引き続き、他の要素（Neighborhoodなど）と組み合わせて慎重に捜査を続ける。
