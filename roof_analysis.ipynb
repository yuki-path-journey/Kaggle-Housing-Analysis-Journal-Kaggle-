```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x='RoofMatl', y='SalePrice', data=train)
plt.title('Impact pf Roof Material on Sale Price')
plt.xticks(rotation=45)
plt.show()
```
